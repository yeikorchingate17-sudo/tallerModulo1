### 1.1 — Listas y operaciones básicas

Tienes la siguiente lista con la esperanza de vida de 10 países (en años).

````python
# Lista dada — NO modificar
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]
````

**Ejercicio 1.1** — Usando Python puro (sin Pandas ni funciones built-in como `sum()` o `max()`):
- Calcula el **promedio** de la lista usando un bucle `for`
- Encuentra el **valor máximo** usando un bucle `for`
- Encuentra el **valor mínimo** usando un bucle `for`

````python
# Tu código aquí

# Promedio

suma = 0 #variable que sumara todos los valores de la lista

#for para recorrer la lista
for i in esperanzas_de_vida:
    suma += i

#calculo de promedio    
promedio=suma/len(esperanzas_de_vida)
 
 #impresion de promedio   
print(f"La esperanza promedio de vida es: {promedio:.3f}")
#--------------------------------------------------------------------------

# Máximo
valor_maximo=esperanzas_de_vida[0]

for i in esperanzas_de_vida:
   if i>valor_maximo:
        valor_maximo=i

print(f"El valor maximo es: {valor_maximo}")

# Mínimo
valor_minimo=esperanzas_de_vida[0]
for i in esperanzas_de_vida:
    if i<valor_minimo:
        valor_minimo=i

print(f"El valor minimo es: {valor_minimo}")
````

### 1.4 — Listas + Funciones + Bucles (integrador)

**Ejercicio 1.4** — Usa la función creada en 1.3 y la lista del ejercicio 1.1 para:
1. Crear una **nueva lista** llamada `clasificaciones` que contenga la clasificación de cada esperanza de vida
2. Imprimir cada par: `(valor, clasificación)` usando un bucle `for`

````python
# Tu código aquí
#lista esperanzas de vida
esperanzas_de_vida = [72.3, 81.5, 60.1, 78.9, 55.4, 83.2, 67.8, 74.0, 49.3, 79.6]

#funcion
def clasificar_esperanza_vida (numero_años,):
    if numero_años>=75:
        return "Alta"
    elif numero_años>=60 and  numero_años<=74.9:
        return "Media"
    else:
        return "baja"
    
#lista clasificaciones
clasificaciones = []

#recorrido con for de la lista clasificar esperanza de vida y se guarda el valor en la lista clasificaciones 
for valor in esperanzas_de_vida:
    resultado = clasificar_esperanza_vida(valor)
    clasificaciones.append(resultado)

# imprimir cada valor y clasificacion
print("***Tipos de clasificacion***")
for i in range(len(esperanzas_de_vida)):
    valor = esperanzas_de_vida[i]
    clase = clasificaciones[i]    
    print(f"valor: {valor} -- Clasificacion ={clase} ")
````

### 2.5 — Agrupaciones

🔧 **Ejercicio 2.5** — Usa `groupby()` para:
1. Calcular la **esperanza de vida promedio** por `region`
2. Calcular el **gasto en salud promedio** por `nivel_ingresos`
3. Calcular el **número de países** por región
4. ¿Qué región tiene la mayor esperanza de vida promedio? ¿Y la menor?

````python
# Tu código aquí
#1) agrupamos con groupby por region y sacamos promedio
esperanza_region= df.groupby("region")["esperanza_vida"].mean()
print(f"esperanza de vida promedio por region: \n {esperanza_region}")
print("_"*40)

#2) calculo de salud promedio agrupado por nivel de ingresos
promedio_salud = df.groupby("nivel_ingresos")["gasto_salud_usd"].mean() 
#impresion de resultados de mayor a menor 
print("gasto en salud promedio por nivel de ingreso: ")
print(promedio_salud.sort_values(ascending=False))
print("-"*40)

#3) calculo de de paises por region con groupby
paises_por_region= df.groupby("region")["pais"].count()
print(f"numero de paises por region: {paises_por_region}")
print("-"*40)

#4 region con mayor esperanza de vida y region con menor esperanza de vida promedio
promedio_esperanza = df.groupby("region")["esperanza_vida"].mean()

#hallamos la region con el valor mas alto y mas bajo

#trae el nombre de la region maxima
region_mayor= promedio_esperanza.idxmax()
#trae el valor promedio de la region maxima
valor_mayor= promedio_esperanza.max()

#trae el nombre de la region minima
region_menor= promedio_esperanza.idxmin()
#trae el valor promedio de la region minima
valor_menor= promedio_esperanza.min()
print(f"region con mayor esperanza de vida promedio es: {region_mayor} con {valor_mayor.round(3)}")
print(f"region con menor esperanza de vida promedio: {region_menor} con {valor_menor.round(3)}")
````
